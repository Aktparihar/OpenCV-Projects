```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(blank_img,text="ABCD",fontFace=font,fontScale=5, org=(50,300),color=(255,255,255),thickness=20)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = load_img()
display_img(img)
```

```python
# MO - erosion
kernel = np.ones((5,5),dtype=np.uint8)
result = cv2.erode(img,kernel,iterations=3)
display_img(result)
```

```python
white_noise = np.random.randint(low=0,high=2,size=(600,600))
display_img(white_noise)
```

```python
white_img = white_noise*255
noise_img = white_img+img
display_img(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img,cv2.MORPH_OPEN,kernel)
display_img(opening)
```

```python
img = load_img()
black_img = np.random.randint(low=0,high=2,size=(600,600))
display_img(black_img)
```

```python
black_img = black_img*-255
```

```python
black_noise_img = img + black_img
```

```python
black_noise_img[black_img == -255] = 0
```

```python
display_img(black_noise_img)
```

```python
closing = cv2.morphologyEx(black_noise_img,cv2.MORPH_CLOSE,kernel)
```

```python
display_img(closing)
```

```python
img = load_img()
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
display_img(gradient)
```
