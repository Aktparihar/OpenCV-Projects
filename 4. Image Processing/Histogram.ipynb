```python
gorilla = cv2.imread('../../DATA/gorilla.jpg',0)
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
display_img(gorilla)
```

```python
gorilla.shape
```

```python
hist_values = cv2.calcHist([gorilla],[0],None,[256],[0,256])
plt.plot(hist_values)
```

```python
eq_gorilla = cv2.equalizeHist(gorilla)
display_img(eq_gorilla)
```

```python
color_gorilla = cv2.imread('../../DATA/gorilla.jpg')
color_gorilla = cv2.cvtColor(color_gorilla, cv2.COLOR_BGR2RGB)
```

```python
hsv = cv2.cvtColor(color_gorilla, cv2.COLOR_BGR2HSV)
```

```python
hsv[:,:,2] = cv2.equalizeHist(hsv[:,:,2])
```

```python
eq_color_gorilla = cv2.cvtColor(hsv,cv2.COLOR_HSV2BGR)
display_img(eq_color_gorilla)
```
