```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display(img,cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap=cmap)
```

```python
sep_coin = cv2.imread("../../DATA/pennies.jpg")
display(sep_coin)
```

```python
sep_blur = cv2.medianBlur(sep_coin,9)
display(sep_blur)
```

```python
gray_sep_coin = cv2.cvtColor(sep_blur, cv2.COLOR_BGR2GRAY)
display(gray_sep_coin,cmap='gray')
```

```python
ret, sep_thresh = cv2.threshold(gray_sep_coin,160,255,cv2.THRESH_BINARY_INV)
display(sep_thresh,cmap='gray')
```

```python
contours,hierarchy = cv2.findContours(sep_thresh.copy(),cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
for i in range(len(contours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(sep_coin,contours,i,(255,0,0),10)
display(sep_coin)
```

```python
# Now we use watershed algorithm
```

```python
img = cv2.imread("../../DATA/pennies.jpg")
```

```python
img = cv2.medianBlur(img,35)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
display(thresh)
```

```python
#noise removal
kernel = np.ones((3,3),np.uint8)
kernel
```

```python
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel,iterations=2)
display(opening,cmap='gray')
```

```python
sure_bg = cv2.dilate(opening,kernel,iterations=3)
display(sure_bg,cmap='gray')
```

```python
dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,5)
display(dist_transform,cmap='gray')
```

```python
ret,sure_fg = cv2.threshold(dist_transform,0.7*dist_transform.max(),255,0)
display(sure_fg,cmap='gray')
```

```python
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg,sure_fg)
display(unknown,cmap='gray')
```

```python
ret,markers = cv2.connectedComponents(sure_fg)
markers = markers+1
markers[unknown == 255] = 0
display(markers,cmap='gray')
```

```python
markers = cv2.watershed(img,markers)
display(markers,cmap='gray')
```

```python
contours,hierarchy = cv2.findContours(markers.copy(),cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
for i in range(len(contours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(sep_coin,contours,i,(255,0,0),10)
display(sep_coin)
```
